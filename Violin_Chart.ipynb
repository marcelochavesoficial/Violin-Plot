# Violin Plot


A very useful visualization for understanding the distribution of a variable is the Violin Plot.

Violin Plot.png

Violin Plots are a way of plotting numerical data, and can be considered a combination of the Box plot and the Kernel Density Plot.
In Violin Plot, we can find the same information as a Box plot, such as:

* Median (White point on the plot)

* IQR (Dark bar in the center of the plot)

* Upper and Lower Limits (Violin Extremities)

* Outliers (Points beyond the Limits)

But if Violin plots show the same information as a Box plot, why use Violin Plot?

The answer is that there is an advantage of the Violin over the Box plot. In addition to showing the information mentioned above, Violin Plot also relies on the distribution of the data, which gives us a better picture, in addition to the statistical information mentioned earlier.

I'll show you in practice how to plot and use these graphs.

# The Data

To plot the violin plot, I'm going to use a dataset about dengue in Brazil.

In addition to the number of confirmed dengue cases, the dataset also shows data related to rainfall, average, minimum and maximum temperatures in the municipality. The period covered by the data set is from 1998 to 2015.



```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt

# import data to dataframe
DATA_PATH = "https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/dengue-dataset.csv"
df = pd.read_csv(DATA_PATH)
df.data = pd.to_datetime(df.data)
df.set_index('data', inplace=True)

# see the first 5 entries
df.head()
```

The Violin plot shows the distribution of data, like a histogram, as well as statistical information like a Boxplot. To make this comparison clearer, I will first plot a Histogram.

```python
# plot the average temperature histogram
fig, ax = plt.subplots()

ax.hist('temperatura-media', data=df)
ax.set_title("Average Temperature Histogram")
ax.set_ylabel("Number of occurrences")
ax.set_xlabel("Temperature")

plt.tight_layout()
```

To plot Violin Plot, I'm going to import Seaborn, another powerful visualization library.

Notice that inside the violin plot, there is a structure that resembles the box plot, in addition to being able to see the white dot, marking the median.

The "curves" on the violin represent the distribution.

```python
import seaborn as sns

# defining the theme
sns.set_style('white')

# plot the boxplot for temperatures
fig, ax = plt.subplots(figsize=(15,10))

del df['casos-confirmados']
del df['chuva']

sns.violinplot(data=df, ax=ax)

plt.tight_layout()
```
